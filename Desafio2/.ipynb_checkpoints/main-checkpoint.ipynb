## Questão 1

Qual a diferença entre os quartis (Q1, Q2 e Q3) das variáveis `normal` e `binomial` de `dataframe`? Responda como uma tupla de três elementos arredondados para três casas decimais.

Em outra palavras, sejam `q1_norm`, `q2_norm` e `q3_norm` os quantis da variável `normal` e `q1_binom`, `q2_binom` e `q3_binom` os quantis da variável `binom`, qual a diferença `(q1_norm - q1 binom, q2_norm - q2_binom, q3_norm - q3_binom)`?

```python
def q1():
    normal = np.percentile(dataframe.normal, [25, 50, 75])
    binomial = np.percentile(dataframe.binomial, [25, 50, 75])
    return (round(normal[0]-binomial[0], 3),round(normal[1]-binomial[1], 3),round(normal[2]-binomial[2], 3))
    # Retorne aqui o resultado da questão 1.
    pass
```

## Questão 2

Considere o intervalo $[\bar{x} - s, \bar{x} + s]$, onde $\bar{x}$ é a média amostral e $s$ é o desvio padrão. Qual a probabilidade nesse intervalo, calculada pela função de distribuição acumulada empírica (CDF empírica) da variável `normal`? Responda como uma único escalar arredondado para três casas decimais.

```python
def q2():
    ecdf = ECDF(dataframe.normal)
    probMaior = ecdf(dataframe.normal.mean() + dataframe.normal.std())
    probMenor = ecdf(dataframe.normal.mean() - dataframe.normal.std())
    answer2 = float(round(probMaior-probMenor,3))
    return answer2
    # Retorne aqui o resultado da questão 2.
    pass
```

## Questão 3

Qual é a diferença entre as médias e as variâncias das variáveis `binomial` e `normal`? Responda como uma tupla de dois elementos arredondados para três casas decimais.

Em outras palavras, sejam `m_binom` e `v_binom` a média e a variância da variável `binomial`, e `m_norm` e `v_norm` a média e a variância da variável `normal`. Quais as diferenças `(m_binom - m_norm, v_binom - v_norm)`?

```python
def q3():
    dif_media = round(dataframe.binomial.mean() - dataframe.normal.mean(),3)
    dif_var = round(dataframe.binomial.var() - dataframe.normal.var(), 3)
    answer3 = tuple([dif_media, dif_var])
    return answer3
```

### _Setup_ da parte 2

```python
stars = pd.read_csv("pulsar_stars.csv")

stars.rename({old_name: new_name
              for (old_name, new_name)
              in zip(stars.columns,
                     ["mean_profile", "sd_profile", "kurt_profile", "skew_profile", "mean_curve", "sd_curve", "kurt_curve", "skew_curve", "target"])
             },
             axis=1, inplace=True)

stars.loc[:, "target"] = stars.target.astype(bool)
```

## Inicie sua análise da parte 2 a partir daqui

```python
# Sua análise da parte 2 começa aqui.
stars_false = stars.query('target == 0')
mean_profile = stars_false['mean_profile']
media = mean_profile.mean()
desvio_padrao = mean_profile.std()
false_pulsar_mean_profile_standardized = (mean_profile - media) / desvio_padrao
ecdf = ECDF(false_pulsar_mean_profile_standardized)
quantil1 = sct.norm.ppf(0.80, loc=0, scale=1) 
quantil2 = sct.norm.ppf(0.90, loc=0, scale=1) 
quantil3 = sct.norm.ppf(0.95, loc=0, scale=1)
```

```python
def q4():
    return tuple([
    round(ecdf(quantil1),3),
    round(ecdf(quantil2),3),
    round(ecdf(quantil3),3)])
```
